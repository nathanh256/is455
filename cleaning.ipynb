```python
'''transforms column to int, which is number of days unitl Date'''
def days_to_date(df, cols, date=datetime.date.today()):
  df_copy = df.copy()
  for i in cols:
    df_copy[i] = df_copy[i].apply(lambda x: datetime.datetime.strptime(x[:20],"%Y-%m-%d"))
    df_copy[i] = df_copy[i].apply(lambda x: (date - x).days)
  return df_copy
```

```python
def fix_skew(df, cols):
    df_copy = df.copy()
    for i in cols:
        df_copy[f"{i}_ln"] = np.log(df_copy[i])
    return df_copy
```
